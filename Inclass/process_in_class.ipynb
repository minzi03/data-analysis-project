```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/module_5_auto.csv')
```

```python
df=df._get_numeric_data()
df.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score, cross_validate
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Giả sử x_data và y_data đã được định nghĩa như sau:
x_data = ['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']
y_data = ['price']
```

```python
# Chia dữ liệu thành x_data và y_data
X = df[x_data]
y = df[y_data]
```

```python
# Chia dữ liệu thành tập Train và Test (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Danh sách để lưu trữ MSE cho từng bậc đa thức trên các tập Train, Test
mse_train_cv = []
mse_test_cv = []
```

```python
# Các bậc đa thức
order = [1, 2, 3, 4, 5, 6]
```

```python
for n in order:
    # Khởi tạo và chuyển đổi dữ liệu với các đặc trưng đa thức
    poly_features = PolynomialFeatures(degree=n)
    X_train_poly = poly_features.fit_transform(X_train)
    X_test_poly = poly_features.transform(X_test)
    
    # Tạo mô hình hồi quy tuyến tính
    model = LinearRegression()
    
    # Cross-validation trên tập huấn luyện (Train)
    mse_train = cross_val_score(model, X_train_poly, y_train, cv=5, scoring='neg_mean_squared_error')
    mse_train_cv.append(-np.mean(mse_train))  # Lấy giá trị MSE, cross_val_score trả về giá trị âm
    
    # Huấn luyện mô hình và tính MSE trên tập Test
    model.fit(X_train_poly, y_train)
    y_test_pred = model.predict(X_test_poly)
    mse_test_cv.append(mean_squared_error(y_test, y_test_pred))
```

```python
# Vẽ biểu đồ MSE cho Train và Test (sử dụng cross-validation)
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(order, mse_train_cv, 'b-o', label='MSE Train (CV)')
ax.plot(order, mse_test_cv, 'r-s', label='MSE Test')

ax.set_xlabel('Polynomial Degree')
ax.set_ylabel('Mean Squared Error (MSE)')
ax.set_title('MSE for Train and Test Datasets Using Cross-Validation')
ax.legend()

plt.show()
```
