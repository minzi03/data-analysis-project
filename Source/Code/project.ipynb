Data Cleaning

```python
# Convert date columns to datetime
date_columns = ['date', 'sent_at', 'opened_first_time_at', 'opened_last_time_at', 
                'clicked_first_time_at', 'clicked_last_time_at', 'unsubscribed_at', 
                'hard_bounced_at', 'soft_bounced_at', 'complained_at', 'blocked_at', 
                'purchased_at', 'created_at', 'updated_at']
df[date_columns] = df[date_columns].apply(pd.to_datetime, errors='coerce')
```

```python
# Convert relevant columns to Boolean
boolean_columns = ['is_opened', 'is_clicked', 'is_unsubscribed', 'is_hard_bounced', 
                   'is_soft_bounced', 'is_complained', 'is_blocked', 'is_purchased']
df[boolean_columns] = df[boolean_columns].applymap(lambda x: True if x == 't' else False)
```

Conversion Funnel Analysis
<li>Analyze the typical user journey, from message receipt to open, click, and purchase, to visualize conversion rates at each stage.
<li>This can help identify stages with the highest drop-off rates.

```python
# Calculate conversion rates
conversion_rates = {
    'Open Rate': df['is_opened'].mean(),
    'Click Rate': df[df['is_opened']]['is_clicked'].mean(),
    'Purchase Rate': df[df['is_clicked']]['is_purchased'].mean()
}
```

```python
# Convert to DataFrame for visualization
conversion_df = pd.DataFrame(list(conversion_rates.items()), columns=['Stage', 'Rate'])
```

```python
# Plot conversion funnel
plt.figure(figsize=(10, 6))
sns.barplot(x='Stage', y='Rate', data=conversion_df)
plt.title('Conversion Funnel')
plt.xlabel('Conversion Stage')
plt.ylabel('Rate')
plt.show()
```

```python
# Conversion Funnel by Channel
funnel_by_channel = df.groupby('channel').apply(lambda x: {
    'Open Rate': x['is_opened'].mean(),
    'Click Rate': x[x['is_opened']]['is_clicked'].mean(),
    'Purchase Rate': x[x['is_clicked']]['is_purchased'].mean()
})
funnel_df = pd.DataFrame.from_dict(funnel_by_channel, orient='index')

funnel_df.plot(kind='bar', figsize=(12, 6))
plt.title('Conversion Funnel by Channel')
plt.xlabel('Channel')
plt.ylabel('Rate')
plt.legend(title="Conversion Stages")
plt.show()
```

Event Pathway Analysis
<li>Explore typical pathways (e.g., message -> open -> click -> purchase) using sequence analysis.
<li>This is useful for identifying the most common actions and sequences that lead to conversions.

```python
# Define a new column to represent the user’s pathway
df['pathway'] = df.apply(lambda x: 'Message' +
                                      (' -> Opened' if x['is_opened'] else '') +
                                      (' -> Clicked' if x['is_clicked'] else '') +
                                      (' -> Purchased' if x['is_purchased'] else ''), axis=1)
```

```python
# Calculate pathway frequencies
pathway_counts = df['pathway'].value_counts()

print("Pathway Analysis:\n", pathway_counts)
```

```python
# Extended pathway analysis
df['extended_pathway'] = df.apply(
    lambda x: 'Message' +
              (' -> Opened' if x['is_opened'] else '') +
              (' -> Clicked' if x['is_clicked'] else '') +
              (' -> Purchased' if x['is_purchased'] else '') +
              (' -> Complained' if x['is_complained'] else '') +
              (' -> Unsubscribed' if x['is_unsubscribed'] else ''),
    axis=1
)

extended_pathway_counts = df['extended_pathway'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(y=extended_pathway_counts.index, x=extended_pathway_counts.values)
plt.title('Extended Pathway Analysis')
plt.xlabel('Count')
plt.ylabel('Pathway')
plt.show()
```

```python
# Feature engineering
df['engagement_lag'] = (df['opened_first_time_at'] - df['sent_at']).dt.total_seconds()
client_engagement = df.groupby('client_id')['is_opened'].cumsum().shift()
df['prior_opens'] = client_engagement.fillna(0)
```

# MODELING

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Select features and target variable
features = df[['channel', 'category', 'platform', 'email_provider']]
target = df['is_opened']
```

```python
# Convert categorical variables to numeric
features = pd.get_dummies(features, drop_first=True)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
# Initialize and fit the model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
# Make predictions and evaluate the model
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Model evaluation with cross-validation
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(model, X_train, y_train, cv=5)
print("Cross-Validation Accuracy:", cv_scores.mean())
```
